Chapter 2 : Linear Regression with multiple variables (dataset->hiring.csv)

 Problem: There is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,

2 yr experience, 9 test score, 6 interview score

12 yr experience, 10 test score, 10 interview score

Answer:
53713.86 and 93747.79

```python
!pip install word2number
```

```python
import pandas as pd     #importing required packages
import numpy as np
from sklearn import linear_model
from word2number import w2n
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv("hiring.csv")
d
```

```python
d.experience = d.experience.fillna("zero")
d
```

```python
d.experience = d.experience.apply(w2n.word_to_num)
d
```

```python
import math
mean_test_score=math.floor(d['test_score(out of 10)'].mean())
mean_test_score
```

```python
d['test_score(out of 10)']=d['test_score(out of 10)'].fillna(mean_test_score)
d
```

```python
reg=linear_model.LinearRegression()
reg.fit(d[['experience','test_score(out of 10)','interview_score(out of 10)']],d['salary($)'])
```

```python
reg.predict([[2,9,6]])
reg.predict([[12,10,10]])
```

```python
!pip install joblib
```

```python
import joblib
```

```python
joblib.dump(reg,'model_to_decide_salaries')
```

```python
model_sal=joblib.load('model_to_decide_salaries')
```

```python
model_sal.predict([[12,10,10]])
```
